```python
# üéØ Configure
LOCAL_CREDENTIALS_PATH = "config-local/collab-data-463313-c340ad86b28e.json"
#doing it with full path because I'm too lazy for actual directory management 
CREDENTIALS_PATH = "/Users/inesaitsahalia/Desktop/labeling_data/data_structure/Dima_collab/collab_data/api-keys/collab-data-463313-c340ad86b28e.json"
PROJECT_ID = "collab-data-463313"
SCR_ROOT_DIR = Path("Desktop/labeling_data/data_structure")
SESSION_DIR = Path(SCR_ROOT_DIR/"2024_05_18-session_0001")  # 
BUCKET_NAME = 'fieldwork_curated'
```

```python
# üöÄ Connect to GCS
gcs_client = GCSClient(
    project_id=PROJECT_ID,
    credentials_path=CREDENTIALS_PATH,
)

# ‚úÖ Create bucket
if BUCKET_NAME not in gcs_client.list_buckets():
    gcs_client.create_bucket(BUCKET_NAME)
print(f"Available buckets: {gcs_client.list_buckets()}")

# üì§ Upload session directory
gcs_prefix = f"{BUCKET_NAME}/{SESSION_DIR.name}"

for root, _, files in os.walk(SESSION_DIR):
    for file in files:
        local_path = Path(root) / file
        relative_path = local_path.relative_to(SESSION_DIR)
        gcs_path = f"{gcs_prefix}/{relative_path.as_posix()}"
        gcs_client.upload_file(str(local_path), gcs_path)

# üîç List contents to verify
print("\nUploaded GCS paths:\n")
for path in gcs_client.glob(f"{gcs_prefix}/**"):
    print(path)
```
